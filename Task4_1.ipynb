## **Task-4 SPAM MAIL DETECTION**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
raw_data = pd.read_csv('/content/spam (1).csv',encoding=('ISO-8859-1'))
raw_data
```

```python
raw_data.head()
```

```python
raw_data.shape
```

```python
mail_data = raw_data.where((pd.notnull(raw_data)),'')
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['v1'] == 'spam','v1',]=0
mail_data.loc[mail_data['v1'] == 'ham','v1',]=1
```

```python
mail_data.head()
```

```python
x = mail_data['v2']
y = mail_data['v1']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features, y_train)
```

```python
prediction_training_data = model.predict(x_train_features)
accuracy = accuracy_score(y_train, prediction_training_data)
```

```python
print('Accuracy on training data = ', accuracy)
```

```python
prediction_test_data = model.predict(x_test_features)
accuracy = accuracy_score(y_test, prediction_test_data)
```

```python
print('Accuracy on test data = ', accuracy)
```

```python
input_mail = ["Nah I don't think he goes to usf, he lives around here though"]

# text to features
input_data_features = feature_extraction.transform(input_mail)

# prediction
prediction = model.predict(input_data_features)

print(prediction)

if (prediction[0]==1):
    print('It is a Ham mail !')


else:
    print('It is a Spam mail !')
```
